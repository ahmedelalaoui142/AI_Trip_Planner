```python
from langchain_groq import ChatGroq
```

```python
# Get API key from environment variable (set it in your .env file)
# Note: In production, store this in .env file and never commit it to git
GROQ_API_KEY = os.getenv("GROQ_API_KEY")
```

```python
# Workaround for SSL certificate issues on macOS with pyenv
# Create a custom httpx client with SSL verification disabled
import httpx
from groq import Groq
import warnings
warnings.filterwarnings('ignore', message='Unverified HTTPS request')

# Create httpx client that skips SSL verification (development only!)
http_client = httpx.Client(verify=False)

# Create Groq client with custom http client  
groq_client = Groq(api_key=GROQ_API_KEY, http_client=http_client)

# Initialize ChatGroq with the custom Groq client
# Available models: llama-3.3-70b-versatile, llama-3.1-8b-instant, mixtral-8x7b-32768
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    client=groq_client.chat.completions
)
```

```python
# Test the LLM
response = llm.invoke("hi")
print(response.content)
```
